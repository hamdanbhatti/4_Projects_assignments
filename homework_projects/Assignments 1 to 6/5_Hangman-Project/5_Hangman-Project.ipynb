```python
import random

print("üéÆ Welcome to Hangman Game!\n")

# Possible words list
words = ["python", "programming", "developer", "hangman", "computer", "game"]

# Computer chooses a random word
word = random.choice(words)
word_letters = set(word)  # unique letters
guessed_letters = set()   # user guessed letters
lives = 6  # number of wrong attempts allowed

while len(word_letters) > 0 and lives > 0:
    # Show current progress
    current_word = [letter if letter in guessed_letters else "_" for letter in word]
    print("Word:", " ".join(current_word))
    print(f"Lives left: {lives}")
    print(f"Guessed letters: {', '.join(sorted(guessed_letters))}\n")

    # User guess
    guess = input("Guess a letter: ").lower()

    if guess in guessed_letters:
        print("‚ö†Ô∏è You already guessed that letter.\n")
        continue

    if guess in word_letters:
        guessed_letters.add(guess)
        word_letters.remove(guess)
        print("‚úÖ Correct guess!\n")
    else:
        lives -= 1
        guessed_letters.add(guess)
        print("‚ùå Wrong guess!\n")

# Game Over / Win
if lives == 0:
    print(f"üò¢ Game Over! The word was: {word}")
else:
    print(f"üéâ Congratulations! You guessed the word: {word}")
```
